Text classification used a bag of words technique that considered unigrams(single tokens)

VADER is a NLTK module that provides sentiment scores based on words used

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('moviereviews.tsv', sep='\t')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
#remove review with spaces
spaces_index = []
for ind,lab,rev in df.itertuples():
    if type(rev) == str:
        if rev.isspace():
            spaces_index.append(ind)
print(spaces_index)
```

```python
df.drop(spaces_index,inplace = True)
df.info()
```

```python
#use vader to calculate polarity scores and add them in dataframe
```

```python
df['scores'] = df['review'].apply(lambda review: sid.polarity_scores(review))

df.head()
```

```python
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])

df.head()
```

```python
df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')

df.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(df['label'],df['comp_score'])
```

```python
print(classification_report(df['label'],df['comp_score']))
```
